### Database with 5*5 images

shapes : square, triangle, cirle
colors : light red, light blue, light yellow, light green, light purple

dataset size : 200,000

probability to have a symbol at each position : 0.5


```python
# Generating the first DB to test the code
db_path = "data/db0.1.0/db.json"
img_dir = "data/db0.1.0/img/"
```

```python
from dbimg import load_db, generate_uuid

db = load_db(db_path)
```

```python
# Number of images generated
NBGEN = 200000

# Grid division of each image
X_DIVISIONS = 5
Y_DIVISIONS = 5

# Size of the images in pixels
img_size = (1000, 1000)

# Probability to generate a geometrical shape at each position in the grid
SHAPE_PROB = 0.5

# Define available shapes
SHAPES = ['circle', 'square', 'triangle']
COLORS  = ["#F86C62", "#DFB5F8", "#7AB0CD", "#F4D67B", "#87C09C"]
```

```python
import numpy as np
from dbimg import generate_uuid
import os
import tqdm

for genidx in tqdm.tqdm(range(NBGEN)):
    content = []
    for i in range(X_DIVISIONS):
        for j in range(Y_DIVISIONS):
            if np.random.random() < SHAPE_PROB:
                content.append({
                    "shape": np.random.choice(SHAPES),
                    "pos": (i, j),
                    "color": np.random.choice(COLORS)
                })

    imgid = generate_uuid()
    db[imgid] = {
        "path": os.path.join(img_dir, imgid + ".png"),
        "division" : (X_DIVISIONS, Y_DIVISIONS),
        "size": img_size,
        "content": content
    }
```

```python
from genimg import gen_img_and_save_db
gen_img_and_save_db(db, db_path, overwrite=False, n_jobs=18)
```

### Create dataset with blue diagonal rule

```python
csv_path = "data/db0.1.0/bluediagonal.csv"
sample_img_path = "data/db0.1.0/bluediagonal"

def blue_diagonal_rule(img_content):
    symbols_in_diagonal = False
    for c in img_content:
        if c["pos"][0] == c["pos"][1]:
            symbols_in_diagonal = True
            if c["color"] != "#7AB0CD":
                return False
    return symbols_in_diagonal
```
