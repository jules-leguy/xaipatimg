```python
import os

data_file = os.environ["DATA"] + "webinterfaces/exp02/filtered_data.pkl"
```

```python
import pandas as pd

df = pd.read_pickle(data_file)
```

```python
from xaipatimg.analysis.dataframes import convert_participants_df_to_answers_times_df

answer_times_df = convert_participants_df_to_answers_times_df(df)
```

```python
from xaipatimg.analysis.dataframes import convert_participants_df_to_within_measures_df

measures_df = convert_participants_df_to_within_measures_df(df)
```

```python
from xaipatimg.analysis.plots import plot_histogram_with_density_by_subgroup

plot_histogram_with_density_by_subgroup(answer_times_df, "XAI condition", "Answer time", "Difficulty/Time pressure", title="Answer time")
```

```python
only_H_HAI_mask_answer_times = ((answer_times_df["XAI condition"] == "H" ) | (answer_times_df["XAI condition"] == "H+AI"))
only_H_HAI_mask_measures = ((measures_df["XAI condition"] == "H" ) | (measures_df["XAI condition"] == "H+AI"))
```

```python
answer_times_df_start_easy_mild = answer_times_df[(answer_times_df["tasks_order"].str[0] == "easy_mild") & only_H_HAI_mask_answer_times]
measures_df_start_easy_mild = measures_df[(measures_df["tasks_order"].str[0] == "easy_mild") & only_H_HAI_mask_measures]
```

```python
answer_times_df_start_easy_strong = answer_times_df[(answer_times_df["tasks_order"].str[0] == "easy_strong") & only_H_HAI_mask_answer_times]
measures_df_start_easy_strong = measures_df[(measures_df["tasks_order"].str[0] == "easy_strong") & only_H_HAI_mask_measures]
```

```python
answer_times_df_start_hard_mild =answer_times_df[(answer_times_df["tasks_order"].str[0] == "hard_mild") & only_H_HAI_mask_answer_times]
measures_df_start_hard_mild = measures_df[(measures_df["tasks_order"].str[0] == "hard_mild") & only_H_HAI_mask_measures]
```

```python
answer_times_df_start_hard_strong = answer_times_df[(answer_times_df["tasks_order"].str[0] == "hard_strong") & only_H_HAI_mask_answer_times]
measures_df_start_hard_strong = measures_df[(measures_df["tasks_order"].str[0] == "hard_strong") & only_H_HAI_mask_measures]
```
