```python
# ============================================================
# START-CONDITION ART-MIXEDLM ANALYSIS (PER XAI CONDITION)
# DV ~ difficulty_start * time_pressure_start
# ============================================================

import pandas as pd
import numpy as np
import ast
from pathlib import Path
import statsmodels.formula.api as smf

# ----------------------------
# SETTINGS
# ----------------------------

OUTPUT_ROOT = Path("start_condition_ART_MixedLM_per_condition")
OUTPUT_ROOT.mkdir(exist_ok=True)

TASK_TYPES = ["easy_mild", "easy_strong", "hard_mild", "hard_strong"]

DV_FULL_LIST = [
    "score_full",
    "reliance_full",
    "overreliance_full",
    "underreliance_full",
    "trust_full",
    "cogload_full",
]

CONDITIONS = ["H", "H+AI"]  # only these conditions

# ============================================================
# PARSE TASK ORDER
# ============================================================

def parse_tasks(x):
    if isinstance(x, str):
        return ast.literal_eval(x)
    return x

df["tasks_order_list"] = df["tasks_order"].apply(parse_tasks)
df["first_task"] = df["tasks_order_list"].str[0]

# ============================================================
# CREATE START-CONDITION VARIABLES
# ============================================================

df["difficulty_start"] = df["first_task"].apply(
    lambda t: "easy" if "easy" in t else "hard"
)

df["time_pressure_start"] = df["first_task"].apply(
    lambda t: "mild" if "mild" in t else "strong"
)

# ============================================================
# BUILD AGGREGATED TRUST AND COGLOAD
# ============================================================

trust_cols = [
    "trust_easy_mild",
    "trust_easy_strong",
    "trust_hard_mild",
    "trust_hard_strong",
]

cogload_cols = [
    "cogload_easy_mild",
    "cogload_easy_strong",
    "cogload_hard_mild",
    "cogload_hard_strong",
]

df["trust_full"] = df[trust_cols].mean(axis=1)
df["cogload_full"] = df[cogload_cols].mean(axis=1)

# ============================================================
# ART TRANSFORM FUNCTION
# ============================================================

def art_transform_between(df, dv):
    """Aligned rank transform for between-subject factors"""
    grand_mean = df[dv].mean()
    cell_means = (
        df.groupby(["difficulty_start", "time_pressure_start"])[dv]
        .transform("mean")
    )
    aligned = df[dv] - cell_means + grand_mean
    df[f"{dv}_aligned"] = aligned
    df[f"{dv}_rank"] = aligned.rank()
    return df

# ============================================================
# ANALYSIS LOOP PER CONDITION
# ============================================================

for cond in CONDITIONS:

    print(f"\n========== CONDITION: {cond} ==========")

    df_c = df[df["xai_condition"] == cond].copy()
    cond_dir = OUTPUT_ROOT / cond
    cond_dir.mkdir(exist_ok=True)

    for dv in DV_FULL_LIST:

        # No trust in H condition
        if dv == "trust_full" and cond == "H":
            continue

        print(f"\nDV: {dv}")

        df_dv = df_c.dropna(subset=[dv]).copy()
        if len(df_dv) < 5:
            print(f"Not enough data for {dv} in condition {cond}, skipping.")
            continue

        # --------------------------------------------------------
        # ART transformation
        # --------------------------------------------------------
        df_art = art_transform_between(df_dv, dv)

        # --------------------------------------------------------
        # MixedLM
        # --------------------------------------------------------
        formula = f"{dv}_rank ~ difficulty_start * time_pressure_start"

        model = smf.mixedlm(
            formula,
            df_art,
            groups=df_art.index  # one row per participant
        )

        result = model.fit()
        print(result.summary())

        # Save summary
        with open(cond_dir / f"ART_MixedLM_{dv}.txt", "w") as f:
            f.write(result.summary().as_text())

print("\nAnalysis complete for H and H+AI conditions only.")
```
