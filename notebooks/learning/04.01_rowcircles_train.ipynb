```python
import os

db_dir = os.environ["DATA"] + "PatImgXAI_data/db1.0.0/"

train_dataset_filename = "rowcircles_train.csv"
test_dataset_filename = "rowcircles_test.csv"
valid_dataset_filename = "rowcircles_valid.csv"
model_dir = os.environ["DATA"] + "models/db1.0.0/rowcircles/"
classification_dir = os.environ["DATA"] + "models/db1.0.0/rowcircles/test_classif/"

device = "cuda:0"
```

```python
from xaipatimg.ml.learning import train_resnet18_model

train_resnet18_model(db_dir, train_dataset_filename, valid_dataset_filename, model_dir, target_accuracy=0.8, device=device, interval_batch=1)
```

```python
from xaipatimg.ml.learning import compute_resnet18_model_scores

compute_resnet18_model_scores(db_dir, train_dataset_filename, test_dataset_filename, valid_dataset_filename, model_dir, device=device)
```
