```python
import csv

def load_json(json_path, set_id=None):
    with open(json_path) as json_data:
        d = json.load(json_data)
        json_data.close()

    if set_id is not None:
        d["view_id"] = set_id

    return d

def load_task_csv(csv_path):
    return list(csv.DictReader(open(csv_path), delimiter=','))
```

```python
def gen_page_new_rule(rule_text, pos_example_path, neg_example_path, training=False, is_new_question=True, view_id=None, time_limit=-1):
    return {
        "type": "p-instruction",
        "view_id": view_id,
        "title": f"<h3>New task ({("training" if training else "evaluated")})<h3>",
        "body_text": f"{"⚠️ New question ⚠️<br/>" if is_new_question else ""}<br/><div class='text-start'><p>The question for the task is : <b>{rule_text}</b> </p> <p> {"For this task, there is no time limit to answer the questions." if time_limit==-1 else f"For this task, you will have <b>{time_limit} seconds</b> at most to answer each question."}</p><br/><br/><div class='container-fluid text-center'> <div class='row justify-content-center'> <div class='col-12 col-md-5 d-flex flex-column align-items-center'> <img src='{pos_example_path}' class='img-fluid w-70' alt='Example of image with answer Yes'> Example of image with answer <b>Yes</b> </div> <div class='col-12 col-md-5 d-flex flex-column align-items-center'> <img src='{neg_example_path}' class='img-fluid w-100' alt='Example of image with answer No'> Example of image with answer <b>No</b> </div> </div> </div>",
        "with_button": True,
        "button_text": f"Start the task ({("training" if training else "evaluated")})",
    }

def get_column_csv_xai(condition):
    if condition == "XAI_SHAP":
        return "xai_shap"
    elif condition == "XAI_CF":
        return "xai_cf"
    # elif condition == "XAI_LLM":
    #     return "xai_llm"
    elif condition == "AI":
        return "AI"
    else:
        return ""

def generate_task_instances(rule_csv_path, condition):

    output_instances_d = []

    csv_content = load_task_csv(rule_csv_path)

    for idx in range(len(csv_content)):

        model_d = {
            "is_image": True,
            "label": os.path.join("assets/", csv_content[idx][get_column_csv_xai(condition)]),
            "title": "AI"
        } if "ai" in condition.lower() else {}

        input_d = {
            "is_image": True,
            "label": os.path.join("assets", csv_content[idx]["source"]),
            "title": "Image to analyze"
        }

        curr_instance_d = {
                "input": input_d,
                "expected" : 0 if csv_content[idx]["target"] == "1" else 1,
        }
        if model_d:
            curr_instance_d["model"] = model_d

        output_instances_d.append(
            curr_instance_d
        )

    return output_instances_d

def generate_task_page(rule_txt, rule_csv_path, condition, time_limit, activate_feedback_answer=False, is_training=False, view_id=None):

    return {
      "type": "p-task",
      "view_id": view_id,
      # "title": "Answer the question related to the image to analyze, before the timer ends." if time_limit != -1 else "Answer the question related to the image to analyze.",
      "desc" : f"<div style='font-size:1.2em'> {"Training task" if is_training else "Task"} question: <b>{rule_txt} </b></div>",
      "show_progression_bar": True,
      "randomize": True,
      "timer": time_limit,
      "time_exceeded_timer": 3,
      "feedback_answer_activated": activate_feedback_answer,
      "feedback_answer_correct": "This is the right answer",
      "feedback_answer_wrong": "This is not the right answer",
      "feedback_answer_show_expected": False,
      "feedback_answer_expected_text": "Expected answer : ",
      "instances": generate_task_instances(rule_csv_path, condition),
      "question": {
        "type": "button",
        "primary_text" : "<div style='font-size:0.8em'>Select the answer using the buttons below, before the timer ends.</div>" if time_limit != -1 else "<div style='font-size:0.8em'>Select the answer using the buttons below.</div>",
        # "secondary_text": f"<div style='font-size:1.2em'> <b>{rule_txt} </b></div><br>",
        "answers": [
          "Yes",
          "No"
        ],
        "options": {
            "css_class_colors": ['btn-success', 'btn-danger']
        }
      }
    }

def generate_page_score(final_score=False, view_id=None):
    return {
        "type": 'p-instruction',
        "view_id": view_id,
        "title": f"{'<h3>Final score</h3>' if final_score else '<h3>Task end</h3>'}",
        "body_text" : f"{'During the main experimental phase, you achieved a score of' if final_score else 'You achieved a score of'}",
        "with_button": True,
        "button_text": f"{'Leave' if final_score else 'Next'}",
        "score" : True
    }
```

```python
def generate_instructions_page(title, content, button_text=None, view_id=None):
    return {
      "type": "p-instruction",
      "view_id": view_id,
      "title": f"<h3>{title}</h3>",
      "body_text": content,
      "button_text": button_text if button_text else "Next",
      "with_button": True
    }
```

```python
def generate_task(task, condition, time, display_is_training, activate_feedback_answer, score_page, is_new_question, view_id_prefix):

    task_d = load_json(os.path.join(interface_dir, "res", "tasks", task+".json"))
    rule_text = task_d["question"]
    pos_example_path = task_d["pos_example"]
    neg_example_path = task_d["neg_example"]
    rule_csv_path = os.path.join(interface_dir, "res", "tasks", task+"_content.csv")

    task_json_list = []
    task_json_list.append(gen_page_new_rule(rule_text, pos_example_path, neg_example_path, display_is_training, is_new_question, view_id=view_id_prefix+"_rule",
                                            time_limit=time))
    task_json_list.append(generate_task_page(rule_text, rule_csv_path, condition, time, activate_feedback_answer=activate_feedback_answer,
                                             is_training=display_is_training, view_id=view_id_prefix+"_task"))
    if score_page:
        task_json_list.append(generate_page_score())
    return task_json_list

def generate_end_of_session_questionnaire(condition, view_id_prefix):
    task_json_list = []

    if "xai" in condition.lower():
        task_json_list.append(load_json(JSON_PATH_TASK_END_QUEST_XAI, set_id=view_id_prefix + "_reliance"))

    elif "ai" in condition.lower():
        task_json_list.append(load_json(JSON_PATH_TASK_END_QUEST_AI, set_id=view_id_prefix + "_reliance"))

    task_json_list.append(load_json(JSON_PATH_TASK_END_COGLOAD, set_id=view_id_prefix + "_cogload"))

    return task_json_list

def create_protocol_json(condition, discovery_training_tasks, easy_training_tasks, difficult_training_tasks, easy_tasks, difficult_tasks, time_mild, time_strong,
                         time_training):
    """
    :param condition: Either "H", "AI", "XAI_SHAP", "XAI_CF", "XAI_LLM"
    :return:
    """
    full_list = []

    # full_list.append(load_json(JSON_PATH_CONSENT_FORM))
    full_list.append(load_json(JSON_PATH_INTRO_FORM))
    full_list.append(load_json(JSON_PATH_INTRO_DISCOVERY_FORM))
    full_list.extend(generate_task(discovery_training_tasks[0], "H", -1, display_is_training=True, activate_feedback_answer=True, score_page=True,
                                   is_new_question=False, view_id_prefix="intro_discovery"))
    full_list.append(load_json(JSON_PATH_INTRO_FURTHER_INFO_FORM))

    if condition == "AI":
        full_list.append(load_json(JSON_PATH_INSTRUCTIONS_AI_FORM))
    elif condition == "XAI_SHAP":
        full_list.append(load_json(JSON_PATH_INSTRUCTIONS_XAI_SHAP_FORM))
    elif condition == "XAI_CF":
        full_list.append(load_json(JSON_PATH_INSTRUCTIONS_XAI_CF_FORM))
    elif condition == "XAI_LLM":
        full_list.append(load_json(JSON_PATH_INSTRUCTIONS_XAI_LLM_FORM))

    if "AI" in condition:
        full_list.append(load_json(JSON_PATH_INTRO_ACCLIM_1_AI))
    else:
        full_list.append(load_json(JSON_PATH_INTRO_ACCLIM_1))
    full_list.extend(generate_task(discovery_training_tasks[1], condition, -1, display_is_training=True, activate_feedback_answer=True, score_page=True,
                                   is_new_question=False, view_id_prefix="intro_acclim_1"))

    if "AI" in condition:
        full_list.append(load_json(JSON_PATH_INTRO_MAIN_TRAINING_AI))
    else:
        full_list.append(load_json(JSON_PATH_INTRO_MAIN_TRAINING))
    full_list.extend(generate_task(easy_training_tasks[0], condition, time_training, display_is_training=True, activate_feedback_answer=False, score_page=True,
                                   is_new_question=True, view_id_prefix="intro_training_1"))
    full_list.extend(generate_task(difficult_training_tasks[0], condition, time_training, display_is_training=True, activate_feedback_answer=False, score_page=True,
                                   is_new_question=True, view_id_prefix="intro_training_3"))


    # Session 1
    full_list.append(load_json(JSON_PATH_MAIN_EXP_START))
    full_list.append(load_json(JSON_PATH_SESSION_1_INTRO))
    full_list.extend(generate_task(easy_tasks[0], condition, time_mild, display_is_training=False, activate_feedback_answer=False, score_page=False,
                                   is_new_question=True, view_id_prefix="mainexp_easy_mild_1"))
    full_list.extend(generate_task(easy_tasks[1], condition, time_mild, display_is_training=False, activate_feedback_answer=False, score_page=False,
                                   is_new_question=True, view_id_prefix="mainexp_easy_mild_2"))
    full_list.extend(generate_end_of_session_questionnaire(condition, view_id_prefix="mainexp_easy_mild"))
    full_list.append(load_json(JSON_PATH_SESSION_1_PART2))
    full_list.extend(generate_task(easy_tasks[2], condition, time_strong, display_is_training=False, activate_feedback_answer=False, score_page=False,
                                   is_new_question=True, view_id_prefix="mainexp_easy_strong_1"))
    full_list.extend(generate_task(easy_tasks[3], condition, time_strong, display_is_training=False, activate_feedback_answer=False, score_page=False,
                                   is_new_question=True, view_id_prefix="mainexp_easy_strong_2"))
    full_list.extend(generate_end_of_session_questionnaire(condition, view_id_prefix="mainexp_easy_strong"))
    full_list.append(load_json(JSON_PATH_SESSION_1_END))

    # Session 2
    full_list.append(load_json(JSON_PATH_SESSION_2_INTRO))
    full_list.extend(generate_task(difficult_tasks[0], condition, time_mild, display_is_training=False, activate_feedback_answer=False, score_page=False,
                                   is_new_question=True, view_id_prefix="mainexp_hard_mild_1"))
    full_list.extend(generate_task(difficult_tasks[1], condition, time_mild, display_is_training=False, activate_feedback_answer=False, score_page=False,
                                   is_new_question=True, view_id_prefix="mainexp_hard_mild_2"))
    full_list.extend(generate_end_of_session_questionnaire(condition, view_id_prefix="mainexp_hard_mild"))
    full_list.append(load_json(JSON_PATH_SESSION_2_PART2))
    full_list.extend(generate_task(difficult_tasks[2], condition, time_strong, display_is_training=False, activate_feedback_answer=False, score_page=False,
                                   is_new_question=True, view_id_prefix="mainexp_hard_strong_1"))
    full_list.extend(generate_task(difficult_tasks[3], condition, time_strong, display_is_training=False, activate_feedback_answer=False, score_page=False,
                                   is_new_question=True, view_id_prefix="mainexp_hard_strong_2"))
    full_list.extend(generate_end_of_session_questionnaire(condition, view_id_prefix="mainexp_hard_strong"))
    full_list.append(load_json(JSON_PATH_SESSION_2_END))

    full_list.append(load_json(JSON_PATH_FINAL_QUEST_INTRO))
    full_list.append(load_json(JSON_PATH_FINAL_QUEST_NFC, set_id="closing_quest_nfc"))
    if "XAI" in condition:
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_TRUST_XAI_1, set_id="closing_quest_trust_1"))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_TRUST_XAI_2, set_id="closing_quest_trust_2"))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_JUSTIF_USE_XAI, set_id="closing_quest_use_xai"))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_STRATEGY, set_id="closing_quest_strategy"))

    elif "AI" in condition:
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_TRUST_AI_1, set_id="closing_quest_trust_1"))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_TRUST_AI_2, set_id="closing_quest_trust_2"))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_STRATEGY, set_id="closing_quest_strategy"))

    full_list.append(load_json(JSON_PATH_END_SCORE))
    full_list.append(generate_page_score(final_score=True))
    full_list.append(load_json(JSON_PATH_END_LEAVE))
    return full_list
```

```python
import shutil
import json

def create_users_json(conditions, users):
    output_dict = {}

    for condition in conditions:
        for user in users[condition]:
            output_dict[user] = {"roles": "user", "protocol": condition+".json"}

    return output_dict

def deploy(output_json_dir, output_assets_dir, users):

    protocols_path = os.path.join(output_json_dir, "protocols")
    users_path = os.path.join(output_json_dir, "users.json")
    assets_output_path = os.path.join(output_assets_dir, "res")

    # Removing former installation
    shutil.rmtree(protocols_path, ignore_errors=True)
    try:
        os.remove(users_path)
    except FileNotFoundError:
        pass

    os.makedirs(protocols_path)

    for condition in CONDITIONS:
        # Writing of JSON protocol files
        protocol_json = create_protocol_json(condition, DISCOVERY_TRAINING_TASKS, EASY_TRAINING_TASKS, DIFFICULT_TRAINING_TASKS, EASY_TASKS, DIFFICULT_TASKS,
                                             TIME_MILD, TIME_STRONG, TIME_TRAINING)
        with open(os.path.join(protocols_path, condition+".json"), 'w') as f:
            json.dump(protocol_json, f, indent=4)

    # Writing users json file
    users_json = create_users_json(CONDITIONS, users)
    with open(users_path, 'w') as f:
        json.dump(users_json, f, indent=4)

    # Copy assets
    shutil.rmtree(assets_output_path, ignore_errors=True)
    shutil.copytree(os.path.join(interface_dir, "res"), assets_output_path)
```
