```python
from xaipatimg.datagen.dbimg import load_db

db_patterns = load_db(db_patterns_dir)
```

```python
import numpy as np

pattern_3sym_2col_keys = []

# Extracting list of patterns that contain 3 symbols of 3 different shapes and 2 different colors. The two items of the same color cannot be
# on a diagonal.
for k, v in db_patterns.items():
    if len(v["content"]) == 3:
        img_col_d = {}
        img_shape_d = {}
        color_matrix = np.full((2,2), "", dtype="U100")
        for entry in v["content"]:
            img_col_d[entry["color"]] = True
            img_shape_d[entry["shape"]] = True
            color_matrix[entry["pos"][0]][entry["pos"][1]] = entry["color"]

        same_color_on_diagonal = color_matrix[0][0] == color_matrix[1][1] or color_matrix[0][1] == color_matrix[1][0]

        if len(img_col_d.keys()) == 2 and len(img_shape_d.keys()) == 3 and not same_color_on_diagonal:
            pattern_3sym_2col_keys.append(k)
```

```python
import csv

def load_json(json_path, set_id=None):
    with open(json_path) as json_data:
        d = json.load(json_data)
        json_data.close()

    if set_id is not None:
        d["view_id"] = set_id

    return d

def load_task_csv(csv_path):
    return list(csv.DictReader(open(csv_path), delimiter=','))
```

```python
def gen_page_new_rule(rule_text, pos_example_path, neg_example_path, pattern_path=None, training=False, is_new_question=True, view_id=None, time_limit=-1):

    pattern_img_html = f"<div class='container-fluid text-center'> <div class='row justify-content-center'> <div class='col-10 col-md-2 d-flex flex-column align-items-center'><img src='{pattern_path}' class='img-fluid' alt='Pattern to search for'> </div> </div> </div><p>" if pattern_path is not None else ""

    return {
        "type": "p-instruction",
        "view_id": view_id,
        "title": f"<h3>New task ({("training" if training else "evaluated")})<h3>",
        "body_text": f"{" New question <br/>" if is_new_question else ""}<br/><div class='text-start'><p>The question for the task is : <b>{rule_text}</b> </p>{pattern_img_html} {"For this task, there is no time limit to answer the questions." if time_limit==-1 else f"For this task, you will have <b>{time_limit} seconds</b> at most to answer each question."}</p><br/><br/><div class='container-fluid text-center'> <div class='row justify-content-center'> <div class='col-12 col-md-5 d-flex flex-column align-items-center'> <img src='{pos_example_path}' class='img-fluid w-70' alt='Example of image with answer Yes'> Example of image with answer <b>Yes</b> </div> <div class='col-12 col-md-5 d-flex flex-column align-items-center'> <img src='{neg_example_path}' class='img-fluid w-100' alt='Example of image with answer No'> Example of image with answer <b>No</b> </div> </div> </div>",
        "with_button": True,
        "button_text": f"Start the task ({("training" if training else "evaluated")})",
    }

def get_column_csv_xai(condition):
    if condition == "XAI_SHAP":
        return "xai_shap"
    elif condition == "XAI_CF":
        return "xai_cf"
    # elif condition == "XAI_LLM":
    #     return "xai_llm"
    elif condition == "AI":
        return "AI"
    else:
        return ""

def generate_task_instances(rule_csv_path, condition):

    output_instances_d = []

    csv_content = load_task_csv(rule_csv_path)

    for idx in range(len(csv_content)):

        expl_d = [{
            "is_image": True,
            "label": os.path.join("assets/", csv_content[idx][get_column_csv_xai(condition)]),
            "title": "AI"
        }] if "ai" in condition.lower() else []

        input_d = {
            "is_image": True,
            "label": os.path.join("assets", csv_content[idx]["source"]),
            "title": "Image to analyze"
        }

        curr_instance_d = {
                "input": input_d,
                "expected" : 0 if csv_content[idx]["target"] == "1" else 1,
        }
        if expl_d:
            curr_instance_d["explanations"] = expl_d

        output_instances_d.append(
            curr_instance_d
        )

    return output_instances_d

def generate_task_page(rule_txt, rule_csv_path, condition, time_limit, pattern_path, activate_feedback_answer=False, is_training=False, view_id=None):

    pattern_img_html = f"<div class='container-fluid text-center'> <div class='row justify-content-center'> <div class='col-12 col-md-1 d-flex flex-column align-items-center'><img src='{pattern_path}' class='img-fluid' alt='Pattern to search for'> </div> </div> </div>" if pattern_path is not None else ""

    return {
      "type": "p-task",
      "view_id": view_id,
      # "title": "Answer the question related to the image to analyze, before the timer ends." if time_limit != -1 else "Answer the question related to the image to analyze.",
      "desc" : f"<div style='font-size:1.2em'> {"Training task" if is_training else "Task"} question: <b>{rule_txt} </b></div>"
               f"{pattern_img_html}",
      "show_progression_bar": True,
      "randomize": True,
      "timer": time_limit,
      "time_exceeded_timer": 3,
      "feedback_answer_activated": activate_feedback_answer,
      "feedback_answer_correct": "This is the right answer",
      "feedback_answer_wrong": "This is not the right answer",
      "feedback_answer_show_expected": False,
      "feedback_answer_expected_text": "Expected answer : ",
      "instances": generate_task_instances(rule_csv_path, condition),
      "question": {
        "type": "button",
        "primary_text" : "<div style='font-size:0.8em'>Select the answer using the buttons below, before the timer ends.</div>" if time_limit != -1 else "<div style='font-size:0.8em'>Select the answer using the buttons below.</div>",
        # "secondary_text": f"<div style='font-size:1.2em'> <b>{rule_txt} </b></div><br>",
        "answers": [
          "Yes",
          "No"
        ],
        "options": {
            "css_class_colors": ['btn-success', 'btn-danger']
        }
      }
    }

def generate_page_score(final_score=False, view_id=None):
    return {
        "type": 'p-instruction',
        "view_id": view_id,
        "title": f"{'<h3>Final score</h3>' if final_score else '<h3>Task end</h3>'}",
        "body_text" : f"{'During the main experimental phase, you achieved a score of' if final_score else 'You achieved a score of'}",
        "with_button": True,
        "button_text": f"{'Leave' if final_score else 'Next'}",
        "score" : True
    }
```

```python
def generate_instructions_page(title, content, button_text=None, view_id=None):
    return {
      "type": "p-instruction",
      "view_id": view_id,
      "title": f"<h3>{title}</h3>",
      "body_text": content,
      "button_text": button_text if button_text else "Next",
      "with_button": True
    }
```

```python
def generate_task(task, condition, time, display_is_training, activate_feedback_answer, score_page, is_new_question, view_id_prefix, pattern_id):

    task_d = load_json(os.path.join(interface_dir, "res", "tasks", task+".json"))
    rule_text = task_d["question"]
    pos_example_path = task_d["pos_example"]
    neg_example_path = task_d["neg_example"]
    rule_csv_path = os.path.join(interface_dir, "res", "tasks", task+"_content.csv")

    task_json_list = []
    task_json_list.append(gen_page_new_rule(rule_text, pos_example_path, neg_example_path,
                                            os.path.join("assets", "res", "tasks", pattern_id+".png") if pattern_id else None,
                                            display_is_training, is_new_question, view_id=view_id_prefix+"_rule", time_limit=time))
    task_json_list.append(generate_task_page(rule_text, rule_csv_path, condition, time,
                                             pattern_path=os.path.join("assets", "res", "tasks", pattern_id+".png") if pattern_id else None,
                                             activate_feedback_answer=activate_feedback_answer,
                                             is_training=display_is_training, view_id=view_id_prefix+"_task"))
    if score_page:
        task_json_list.append(generate_page_score())
    return task_json_list

def generate_end_of_session_questionnaire(condition, view_id_prefix):
    task_json_list = []

    if "xai" in condition.lower():
        task_json_list.append(load_json(JSON_PATH_TASK_END_QUEST_XAI, set_id=view_id_prefix + "_reliance"))

    elif "ai" in condition.lower():
        task_json_list.append(load_json(JSON_PATH_TASK_END_QUEST_AI, set_id=view_id_prefix + "_reliance"))

    task_json_list.append(load_json(JSON_PATH_TASK_END_COGLOAD, set_id=view_id_prefix + "_cogload"))

    return task_json_list

def generate_session_easy(easy_tasks, condition, time_mild, time_strong, mid_session_json):
    views_list = []
    views_list.extend(generate_task(easy_tasks[0], condition, time_mild, display_is_training=False, activate_feedback_answer=False,
                                   score_page=True, is_new_question=True, view_id_prefix="mainexp_easy5_pat",
                                   pattern_id=rules_data_M[2]["pattern_id"]))
    views_list.extend(generate_task(easy_tasks[1], condition, time_mild, display_is_training=False, activate_feedback_answer=False,
                                   score_page=True, is_new_question=True, view_id_prefix="mainexp_easy2_N", pattern_id=None))

    views_list.extend(generate_end_of_session_questionnaire(condition, view_id_prefix="mainexp_easy_mild"))
    views_list.append(load_json(mid_session_json))

    views_list.extend(generate_task(easy_tasks[2], condition, time_strong, display_is_training=False, activate_feedback_answer=False,
                                   score_page=True, is_new_question=True, view_id_prefix="mainexp_easy6_pat",
                                   pattern_id=rules_data_M[3]["pattern_id"]))
    views_list.extend(generate_task(easy_tasks[3], condition, time_strong, display_is_training=False, activate_feedback_answer=False,
                                   score_page=True, is_new_question=True, view_id_prefix="mainexp_easy4_N", pattern_id=None))

    views_list.extend(generate_end_of_session_questionnaire(condition, view_id_prefix="mainexp_easy_strong"))
    return views_list

def generate_session_hard(difficult_tasks, condition, time_mild, time_strong, mid_session_json):
    views_list = []
    views_list.extend(generate_task(difficult_tasks[0], condition, time_mild, display_is_training=False, activate_feedback_answer=False,
                                   score_page=True, is_new_question=True, view_id_prefix="mainexp_hard5_pat",
                                   pattern_id=rules_data_L[2]["pattern_id"]))
    views_list.extend(generate_task(difficult_tasks[1], condition, time_mild, display_is_training=False, activate_feedback_answer=False, score_page=True, is_new_question=True, view_id_prefix="mainexp_hard2_N", pattern_id=None))

    views_list.extend(generate_end_of_session_questionnaire(condition, view_id_prefix="mainexp_hard_mild"))
    views_list.append(load_json(mid_session_json))

    views_list.extend(generate_task(difficult_tasks[2], condition, time_strong, display_is_training=False, activate_feedback_answer=False,
                                   score_page=True, is_new_question=True, view_id_prefix="mainexp_hard6_pat",
                                   pattern_id=rules_data_L[3]["pattern_id"]))
    views_list.extend(generate_task(difficult_tasks[3], condition, time_strong, display_is_training=False, activate_feedback_answer=False, score_page=True, is_new_question=True, view_id_prefix="mainexp_hard4_N", pattern_id=None))

    views_list.extend(generate_end_of_session_questionnaire(condition, view_id_prefix="mainexp_hard_strong"))
    return views_list

def create_protocol_json(condition, intro_tasks, easy_tasks, difficult_tasks, time_mild, time_strong, time_training):
    """
    :param condition: Either "H", "AI", "XAI_SHAP", "XAI_CF", "XAI_LLM"
    :return:
    """
    full_list = []

    full_list.append(load_json(JSON_PATH_CONSENT_FORM))
    full_list.append(load_json(JSON_PATH_INTRO_DISCOVERY_FORM))
    full_list.append(load_json(JSON_PATH_INTRO_PATTERN_TASK_FORM))
    full_list.extend(generate_task(intro_tasks[0], "H", -1, display_is_training=True, activate_feedback_answer=True, score_page=True,
                                   is_new_question=True, view_id_prefix="intro_discovery1", pattern_id=rules_data_S[2]["pattern_id"]))
    full_list.extend(generate_task(intro_tasks[1], "H", -1, display_is_training=True, activate_feedback_answer=True, score_page=True,
                                   is_new_question=True, view_id_prefix="intro_discovery2", pattern_id=None))
    full_list.append(load_json(JSON_PATH_INTRO_FURTHER_INFO_FORM))

    if condition == "AI":
        full_list.append(load_json(JSON_PATH_INSTRUCTIONS_AI_FORM))
    elif condition == "XAI_SHAP":
        full_list.append(load_json(JSON_PATH_INSTRUCTIONS_XAI_SHAP_FORM))
    elif condition == "XAI_CF":
        full_list.append(load_json(JSON_PATH_INSTRUCTIONS_XAI_CF_FORM))
    # elif condition == "XAI_LLM":
    #     full_list.append(load_json(JSON_PATH_INSTRUCTIONS_XAI_LLM_FORM))

    if "AI" in condition:
        full_list.append(load_json(JSON_PATH_INTRO_MAIN_TRAINING_AI))
    else:
        full_list.append(load_json(JSON_PATH_INTRO_MAIN_TRAINING))

    full_list.extend(generate_task(intro_tasks[2], condition, time_training, display_is_training=True, activate_feedback_answer=False, score_page=True,
                                   is_new_question=True, view_id_prefix="intro_training_1", pattern_id=rules_data_S[3]["pattern_id"]))
    full_list.extend(generate_task(intro_tasks[3], condition, time_training, display_is_training=True, activate_feedback_answer=False, score_page=True,
                                   is_new_question=True, view_id_prefix="intro_training_3", pattern_id=None))

    first_difficulty = np.random.choice(["easy", "hard"])
    print(first_difficulty)

    # Session 1
    full_list.append(load_json(JSON_PATH_MAIN_EXP_START))
    full_list.append(load_json(JSON_PATH_SESSION_1_INTRO))
    if first_difficulty == "easy":
        full_list.extend(generate_session_easy(easy_tasks, condition, time_mild, time_strong, mid_session_json=JSON_PATH_SESSION_1_PART2))
    else:
        full_list.extend(generate_session_hard(difficult_tasks, condition, time_mild, time_strong, mid_session_json=JSON_PATH_SESSION_1_PART2))
    full_list.append(load_json(JSON_PATH_ATTENTIONCHECK1, set_id="attentioncheck_1"))
    full_list.append(load_json(JSON_PATH_SESSION_1_END))

    # Session 2
    full_list.append(load_json(JSON_PATH_SESSION_2_INTRO))
    if first_difficulty == "easy":
        full_list.extend(generate_session_hard(difficult_tasks, condition, time_mild, time_strong, mid_session_json=JSON_PATH_SESSION_2_PART2))
    else:
        full_list.extend(generate_session_easy(easy_tasks, condition, time_mild, time_strong, mid_session_json=JSON_PATH_SESSION_2_PART2))
    full_list.append(load_json(JSON_PATH_SESSION_2_END))


    # full_list.append(load_json(JSON_PATH_FINAL_QUEST_NFC, set_id="closing_quest_nfc"))
    if "XAI" in condition:
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_INTRO))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_TRUST_XAI_1, set_id="closing_quest_trust_1"))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_TRUST_XAI_2, set_id="closing_quest_trust_2"))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_JUSTIF_USE_XAI, set_id="closing_quest_use_xai"))
        full_list.append(load_json(JSON_PATH_ATTENTIONCHECK2, set_id="attentioncheck_2"))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_STRATEGY, set_id="closing_quest_strategy"))

    elif "AI" in condition:
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_INTRO))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_TRUST_AI_1, set_id="closing_quest_trust_1"))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_TRUST_AI_2, set_id="closing_quest_trust_2"))
        full_list.append(load_json(JSON_PATH_ATTENTIONCHECK2))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_STRATEGY, set_id="closing_quest_strategy"))
    else:
        full_list.append(load_json(JSON_PATH_ATTENTIONCHECK2, set_id="attentioncheck2"))
        full_list.append(load_json(JSON_PATH_FINAL_QUEST_NOAI_NOQUEST))
    #
    # full_list.append(load_json(JSON_PATH_END_SCORE))
    # full_list.append(generate_page_score(final_score=True))
    full_list.append(load_json(JSON_PATH_END_LEAVE))
    return full_list
```

```python
import shutil
import json


def deploy(output_json_dir, output_assets_dir):

    protocols_path = os.path.join(output_json_dir, "protocols")
    users_path = os.path.join(output_json_dir, "users.json")
    assets_output_path = os.path.join(output_assets_dir, "res")

    # Removing former installation
    shutil.rmtree(protocols_path, ignore_errors=True)
    try:
        os.remove(users_path)
    except FileNotFoundError:
        pass

    os.makedirs(protocols_path)

    users_json = {}
    for condition in CONDITIONS:
        for protocol_idx in range(N_PROTOCOL_FILES[condition]):
            # Writing of JSON protocol files
            protocol_json = create_protocol_json(condition, INTRO_TASKS, EASY_TASKS, DIFFICULT_TASKS, TIME_MILD, TIME_STRONG, TIME_TRAINING)
            protocol_filename = condition+f"_{protocol_idx}.json"
            with open(os.path.join(protocols_path, protocol_filename), 'w') as f:
                json.dump(protocol_json, f, indent=4)
            users_json[f"{condition}_{protocol_idx}"] = {"roles": "user", "protocol": protocol_filename}

    with open(users_path, 'w') as f:
        json.dump(users_json, f, indent=4)

    # Copy assets
    shutil.rmtree(assets_output_path, ignore_errors=True)
    shutil.copytree(os.path.join(interface_dir, "res"), assets_output_path)
```
